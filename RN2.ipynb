```
# import the necessary packages
from keras_retinanet.utils.image import preprocess_image
from keras_retinanet.utils.image import read_image_bgr
from keras_retinanet.utils.image import resize_image
from keras_retinanet import models
import numpy as np
import argparse
import cv2
```

```
# load the class label mappings
labels = '/content/keras-retinanet/retinanet_classes.csv'
LABELS = open(labels).read().strip().split("\n")
print(LABELS)
LABELS = {int(L.split(",")[1]): L.split(",")[0] for L in LABELS}
print(LABELS)
```

```
# load the model from disk
model_path = '/content/keras-retinanet/output.h5'
model = models.load_model(model_path, backbone_name='resnet50')
```

```
#image = read_image_bgr("/content/7000_3000300154_Page_2.jpg")
image = read_image_bgr("/content/7000_3000300164_Page_1.jpg")
```

```
output = image.copy()
output = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)

image = preprocess_image(image)
(image, scale) = resize_image(image)
image = np.expand_dims(image, axis=0)
```

```
# detect objects in the input image and correct for the image scale
(boxes, scores, labels) = model.predict_on_batch(image)
boxes /= scale
```

```
confidence =  0.5
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# loop over the detections
for (box, score, label) in zip(boxes[0], scores[0], labels[0]):
	# filter out weak detections
	if score < confidence:
		continue

	# convert the bounding box coordinates from floats to integers
	box = box.astype("int")

	# build the label and draw the label + bounding box on the output
	# image
	label = "{}: {:.2f}".format(LABELS[label], score)
	print(label)
	cv2.rectangle(output, (box[0], box[1]), (box[2], box[3]),
		(0, 255, 0), 2)
	print(box[0])
	print(box[1])
	print(box[2])
	print(box[3])
	cv2.putText(output, label, (box[0], box[1] - 10),
		cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# show the output image
#cv2.imshow("Output", output)

plt.figure(figsize=(20, 20))
plt.imshow(output)
plt.show()
#cv2.waitKey(0)
```

```
confidence =  0.5
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from keras_retinanet.utils.visualization import draw_box, draw_caption
from keras_retinanet.utils.colors import label_color

labels_to_names = {0: 'table'}

# visualize detections
for box, score, label in zip(boxes[0], scores[0], labels[0]):
    # scores are sorted so we can break
    if score < 0.5:
        break
        
    color = label_color(label)
    
    b = box.astype(int)
    draw_box(output, b, color=color)
    
    caption = "{} {:.3f}".format(labels_to_names[label], score)
    draw_caption(output, b, caption)
    
plt.figure(figsize=(15, 15))
plt.axis('off')
plt.imshow(output)
plt.show()
```
